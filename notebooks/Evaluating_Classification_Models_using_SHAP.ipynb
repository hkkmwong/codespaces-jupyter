# Pre-Processing



*   Encode categorical variables into numeric
*   Split dataset into training set and test set



```python
from sklearn.model_selection import train_test_split
X,y = shap.datasets.adult()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
X.head()
```

## Model building

### Census income classification with Linear models - Logistic regression  

```python
import shap
from sklearn.linear_model import LogisticRegression
shap.initjs()
```

#### Model Training
Build a logistic regression model to make predictions on the targeted label.


```python
model = LogisticRegression(solver='lbfgs',max_iter=400).fit(X_train, y_train)
y_pred = model.predict(X_test)

explainer = shap.Explainer(model, X, feature_names=X.columns)
shap_values = explainer(X)
```

#### Visualize a single prediction

For local intrepretations, we will focus on the first data sample.

```python
shap.initjs()
shap.plots.force(shap_values[0])
```

From the SHAP values, for this record, we can observe that Education and Capital Gain give the greatest Positive impacts to the prediction. On the other hand, Relationshop and Martial Status give the greatest Negative impacts to the prediction.

#### Summarize the effect of all the features




```python
shap.summary_plot(shap_values, X, plot_type="bar")
```

From the SHAP values, we can observe that these features give the greatest impacts to the prediction.


*   Relationship
*   Marital Status
*   Capital Gain
*   Education 



```python
shap.plots.beeswarm(shap_values)
```

From the SHAP values, we can observe that 

in general:
- Higher value of Relationship, i.e. "Wife" and "Husband" will have higher chance of earning above USD50K/year income
- Lower value of Marital Status, i.e. "Divorced" and "Married-AF-spouse" will have higher chance of earning above USD50K/year income
- People with higher Capital Gain will have higher chance of earning above USD50K/year income
- People with higher Education will have higher chance of earning above USD50K/year income

### Census income classification with Tree-based models - XGBoost
load additional libraries for XGBoost model example

```python
import xgboost
import numpy as np
import matplotlib.pylab as pl
```

#### Load dataset

```python
d_train = xgboost.DMatrix(X_train, label=y_train)
d_test = xgboost.DMatrix(X_test, label=y_test)
```

#### Model Training

```python
params = {
    "eta": 0.01,
    "objective": "binary:logistic",
    "subsample": 0.5,
    "base_score": np.mean(y_train),
    "eval_metric": "logloss"
}
model = xgboost.train(params, d_train, 5000, evals = [(d_test, "test")], verbose_eval=100, early_stopping_rounds=20)
```

#### Visualize a single prediction


```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)
```

```python
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[0,:], X.iloc[0,:])
```

From the SHAP values, for this record, we can observe that Education gives the greatest positive impacts to the prediction. On the other hand, Relationship and Capital Gain give the greatest negative impacts to the prediction.

#### Summarize the effect of all the features


```python
shap.summary_plot(shap_values, X, plot_type="bar")
```

From the SHAP values, we can observe that these features give the greatest impacts to the prediction.


*   Relationship
*   Age
*   Education
*   Capital Gain 



```python
shap.summary_plot(shap_values, X)
```

From the SHAP values, we can observe that 

in general:
- Higher value of Relationship, i.e. "Wife" and "Husband" will have higher chance of earning above USD50K/year income
- Young people will have higher chance of earning below USD50K/year income
- People with higher Education will have higher chance of earning above USD50K/year income

### Census income classification with Neural networks - Keras

load additional libraries for Neural Network model example

```python
from keras.layers import Input, Dense, Flatten, Concatenate, concatenate, Dropout, Lambda
from keras.models import Model
from tensorflow.python.keras.layers.embeddings import Embedding
from tqdm import tqdm
import shap
import numpy as np
import random
```

#### Load dataset

```python
# normalize data (this is important for model convergence)
dtypes = list(zip(X.dtypes.index, map(str, X.dtypes)))

for k,dtype in dtypes:
    if dtype == "float32":
        X[k] -= X[k].mean()
        X[k] /= X[k].std()

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=7)
```

#### Train the model

```python
# build model
input_els = []
encoded_els = []
for k,dtype in dtypes:
    input_els.append(Input(shape=(1,)))
    if dtype == "int8":
        e = Flatten()(Embedding(X_train[k].max()+1, 1)(input_els[-1]))
    else:
        e = input_els[-1]
    encoded_els.append(e)
encoded_els = concatenate(encoded_els)
layer1 = Dropout(0.5)(Dense(100, activation="relu")(encoded_els))
out = Dense(1)(layer1)

# train model
regression = Model(inputs=input_els, outputs=[out])
regression.compile(optimizer="adam", loss='binary_crossentropy')
regression.fit(
    [X_train[k].values for k,t in dtypes],
    y_train,
    epochs=50,
    batch_size=512,
    shuffle=True,
    validation_data=([X_valid[k].values for k,t in dtypes], y_valid)
)
```

#### Explain predictions

SHAP expects model functions to take a 2D numpy array as input, so we define a wrapper function around the original Keras predict function.

```python
def f(X):
    return regression.predict([X[:,i] for i in range(X.shape[1])]).flatten()
```

#### Visualize a single prediction

50 samples from dataset will be used to represent "typical" features values, 500 peterbation samples will be used to estimate the SHAP values for a given prediction.

```python
explainer = shap.KernelExplainer(f, X.iloc[:50,:])
shap_values = explainer.shap_values(X.iloc[0,:], nsamples=500)
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values, X.iloc[0,:])
```

From the SHAP values, for this record, we can observe that Education and Age give the greatest positive impacts to the prediction. On the other hand, Relationship gives the greatest negative impacts to the prediction.
